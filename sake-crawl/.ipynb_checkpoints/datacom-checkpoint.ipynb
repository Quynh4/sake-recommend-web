```python
import urllib
import json
import ssl
import csv
from tqdm import tqdm

ssl._create_default_https_context = ssl._create_unverified_context

area = 1
count = 20

with open('liquor_data.csv', mode='w', encoding='utf-8', newline='') as csv_file:
    fieldnames = [
        'id',
        'brand_intl_name',
        'brand_name',
        'checkin_count',
        'f1', 'f2', 'f3', 'f4', 'f5', 'f6',
        'flavour_tags',
        'intl_name',
        'name',
        'pictures',
        'rank',
        'score',
        'similar_brands',
        'year_month'
    ]

    writer = csv.DictWriter(
        csv_file,
        fieldnames=fieldnames,
        delimiter=',',
        quotechar='"',
        quoting=csv.QUOTE_MINIMAL
    )
    writer.writeheader()

    while area <= 47:
        url = f'https://sakenowa.com/api/v2/brands/ranking?areaId={area}&count={count}'
        response = urllib.request.urlopen(url)
        data = json.loads(response.read())['ranking']

        for item in tqdm(data):
            brand_summary = item['brandSummary']

            year_month = str(item['yearMonth'])
            score = float(item['score'])
            ranking = int(item['rank'])

            brand = brand_summary['brand']
            id_ = int(brand['id'])
            name = str(brand['name'])
            intl_name = brand.get('intlName')

            pictures = [p['url'] for p in brand_summary.get('pictures', [])]
            checkin_count = int(brand_summary['statistics']['checkinCount'])

            similar_brands = [
                b['brand']['name']
                for b in brand_summary.get('similarBrands', [])
            ]

            brand_name = None
            brand_intl_name = None
            brewery = brand.get('brewery')
            if brewery and brewery.get('area'):
                brand_name = brewery['area'].get('name')
                brand_intl_name = brewery['area'].get('intlName')

            f1 = f2 = f3 = f4 = f5 = f6 = None
            if 'simpleFlavorFeature' in brand_summary:
                f = brand_summary['simpleFlavorFeature']
                f1, f2, f3, f4, f5, f6 = f['f1'], f['f2'], f['f3'], f['f4'], f['f5'], f['f6']

            tags = '|'.join(t['tag'] for t in brand_summary.get('flavorTags', []))

            writer.writerow({
                'id': id_,
                'brand_intl_name': brand_intl_name,
                'brand_name': brand_name,
                'checkin_count': checkin_count,
                'f1': f1, 'f2': f2, 'f3': f3,
                'f4': f4, 'f5': f5, 'f6': f6,
                'flavour_tags': tags,
                'intl_name': intl_name,
                'name': name,
                'pictures': '|'.join(pictures),
                'rank': ranking,
                'score': score,
                'similar_brands': '|'.join(similar_brands),
                'year_month': year_month
            })

        area += 1
```
